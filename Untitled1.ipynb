```python
def create_predictions(predictors, core_mumbai, reg):
    train = core_mumbai.loc[:"2020-12-31"]
    test = core_mumbai.loc["2021-01-01":]
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"], predictions)
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    return error, combined
```

```python
core_mumbai["month_max"] = core_mumbai["temp_max"].rolling(30).mean()
```

```python
core_mumbai = core_mumbai.iloc[30:,:].copy()
```

```python
core_mumbai["month_day_max"] = core_mumbai["month_max"] / core_mumbai["temp_max"]
```

```python
core_mumbai["max_min"] = core_mumbai["temp_max"] / core_mumbai["temp_min"]
```

```python
predictors = ["precip","temp_max","temp_min","month_max","month_day_max","max_min"]
```

```python
error, combined = create_predictions(predictors, core_mumbai, reg)
```

```python
core_mumbai["monthly_avg"] = core_mumbai["temp_max"].groupby(core_mumbai.index.month).apply(lambda x: x.expanding(1).mean())
```

```python
core_mumbai["day_of_year_avg"] = core_mumbai["temp_max"].groupby(core_mumbai.index.day_of_year).apply(lambda x: x.expanding(1).mean())
```

```python
error, combined = create_predictions(predictors + ["monthly_avg", "day_of_year_avg"], core_mumbai, reg)
error
```
